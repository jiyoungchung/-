### 회귀분석
- ols 함수로 회귀분석을 실시
- 종속변수 ~ 독립변수의 형태로 모형식

#### 전체 회귀분석 (석식계)

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm
```

```python
# 종속 / 독립변수 
feature_columns = list(Food.columns.difference(["중식계", "석식계","일자"]))

X_data = Food[feature_columns]  # 독립변수
X_data.head()

y_data = Food["석식계"]  # 종속변수
```

```python
# 독립변수에 상수항을 고려
x_data1 = sm.add_constant(X_data, has_constant = "add") 
```

```python
# OLS : 잔차제곱의합을 최소화하는 것
multi_model = sm.OLS(y_data, x_data1)
fitted_multi_model = multi_model.fit()
```

```python
fitted_multi_model.summary()
```

- R^2 : 44.8% -> 46.2%
- 유의하지 않은 변수(P-value > 0.05) : defcnt, new_col(석식), new_col(중식), 강수량(mm), 석식_1_CLEAN, 일조합, 중식_2
